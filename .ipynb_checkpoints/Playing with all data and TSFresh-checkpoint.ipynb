```python
import numpy as np
import pandas as pd
%matplotlib inline
import glob
from functools import reduce
import tsfresh
from tsfresh.utilities.dataframe_functions import impute
from tsfresh import extract_features, extract_relevant_features, select_features
from tsfresh.feature_extraction import ComprehensiveFCParameters
```

```python
from tsfresh.feature_extraction import ComprehensiveFCParameters
```

```python
path = r'/Users/mahnooshsadeghi/Desktop/PTSD/PTSD Data/Windowed data for TSfresh' # use your path
all_files = glob.glob(path + "/*.csv")
```

```python
li=[]
F=[]
```

```python
settings = ComprehensiveFCParameters()
```

```python
for filename in all_files:
    frame = pd.read_csv(filename, index_col=None, header=0)
    stress=frame.groupby('windowno', as_index=False)['ptsd_moment'].max() 
    stress=stress.rename(columns={"ptsd_moment": "stress"})
    result = frame.merge(stress, on=['windowno'])  
    li.append(result)
F = pd.concat(li, axis=0, ignore_index=True)
```

```python
F['windownopar']= F['windowno']*F['windowno']+2*F['participant']*F['participant']+F['windowno']*F['participant']
```
